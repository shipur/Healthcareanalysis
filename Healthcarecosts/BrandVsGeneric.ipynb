```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
drugs = pd.read_csv("drugs.csv")
drugs.head(5)
```

```python
x = np.array([2008,2009,2010,2011,2012,2013,2014,2015,2016])

brand = drugs["Brand Prescription Price Index"]
generic = drugs["Generic Prescription Price Index"]

c, ax1 = plt.subplots(figsize=(7, 7))
b, = ax1.plot(x, brand, 'r-', label="Brand Name Drugs $",)
a, = ax1.plot(x, generic, 'blue', label="Generic Drugs $",)

plt.title('Generic vs Brand Name Drug Costs', fontsize=20)
plt.legend(handles=[b,a])

plt.xlabel('Years')
plt.ylabel('Cost of Prescription Drugs in USD')

plt.grid()
plt.show()
plt.savefig('Generic_Vs_Brand.png')
```
